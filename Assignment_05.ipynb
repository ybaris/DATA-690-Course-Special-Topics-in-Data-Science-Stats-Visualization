## 7. Since there are too many variables to explore, many of which are not interesting. Create a smaller data frame with the following 9 variables:

UNITID
INSTNM
CITY
STABBR
ZIP
ADM_RATE
UGDS
TUITIONFEE_IN
MN_EARN_WNE_P10

```
a = pd.DataFrame()

a["UNITID"] = data["UNITID"]
a["INSTNM"] = data["INSTNM"]
a["CITY"] = data["CITY"]
a["STABBR"] = data["STABBR"]
a["ZIP"] = data["ZIP"]
a["ADM_RATE"] = data["ADM_RATE"]
a["UGDS"] = data["UGDS"]
a["TUITIONFEE_IN"] = data["TUITIONFEE_IN"]
a["MN_EARN_WNE_P10"] = data["MN_EARN_WNE_P10"]

a
# Below could be another way
# data_smaller = data[['UNITID', 'INSTNM', 'CITY', 'STABBR', 'ZIP', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10']]

# data_smaller
```

9. Find out the definition of the above 9 variables and document the definitions in the Jupyter Notebook in a Markdown cell so that the readers of your notebook will know the meaning of these variables.

- UNITID: Unit ID for institution
- INSTNM: Institution name
- CITY: City
- STABBR: State postcode
- ZIP: Zip Code
- ADM_RATE: Admission rate
- UGDS: Enrollment of undergraduate certificate/degree-seeking students
- TUITIONFEE_IN: In-state tuition and fees
- MN_EARN_WNE_P10: Mean earnings of students working and not enrolled 10 years after entry

#### Calculate the min, max, and average in-state tution of the above four institutions.

```
# target_institutes list is created to get desired universities
target_institutes = ['Johns Hopkins University', 'Loyola University Maryland', 'University of Maryland-College Park', 'University of Maryland-Baltimore County']

subset_a = a[a['INSTNM'].isin(target_institutes)]
subset_a
```

```
print(subset_a['TUITIONFEE_IN'].min())
print(subset_a['TUITIONFEE_IN'].max())
print(subset_a['TUITIONFEE_IN'].mean())
```

### 11. Filter the smaller dataset further to only keep colleges from Maryland, Virgina, Washington DC, Delaware, and Pennsylvinia. Save the filtered dataset to a new file named "Five_States_Colleges.csv" for future analysis.

```
# creating a list of desired states
five_states = ['MD', 'VA', 'DC', 'DE', 'PA']

five_state_colleges = a[a['STABBR'].isin(five_states)]

five_state_colleges.head()
```

```
five_state_colleges.to_csv('Five_States_Colleges.csv')
```
